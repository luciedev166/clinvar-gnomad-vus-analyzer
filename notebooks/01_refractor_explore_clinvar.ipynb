```python
from pathlib import Path
import pandas as pd

# set project root
project_root = Path.cwd()

if project_root.name == "notebooks":
    project_root = project_root.parent

project_root
```

```python
# get tsv
tsv = (
    project_root 
    / "data"
    / "raw"
    / "clinvar_download_20260712_233035.tsv"
)
df = pd.read_csv(tsv, sep="\t")
df.head()
```

```python
# filter candidates for VUS missense
candidate_variants = df[
    (df["Genes"] == "ATM") &
    (df["Genes"] == "ATM") &
    (df["Classification"] == "G: Uncertain significance") &
    (df["Molecular consequence"] == "missense variant") &
    (
        df["rs_ID"]
        .astype("string")
        .str.fullmatch(r"rs\d+", na=False)
    )
    &
    (df["GRCh38 Location"].notna())
    &
    (
        df["Review status"]
        .str.startswith("criteria provided", na=False)
    )
    ].copy().reset_index(drop=True)

candidate_variants.head()
```

```python
# output candidates
output_path = (
    project_root
    / "data"
    / "processed"
    / "atm_clinvar_candidates.csv"
)

candidate_variants.to_csv(output_path, index=False)
print("candidate_variants saved to: ", output_path)
```
